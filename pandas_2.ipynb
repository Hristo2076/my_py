```python
movi = pd.read_csv('movies.csv')
movi[:5]
```

```python
rating = pd.read_csv('ratings.csv')
rating.head()
```

```python
movi_rating = movi.merge(rating,left_on='movieId',right_on='movieId')
movi_rating.head()
```

```python
def classification(rating):
  if rating <= 2:
    return 'low rating'
  elif rating <=4:
    return 'medium rating'
  else:
    return 'high rating'
```

```python
movi_rating['class'] = movi_rating['rating'].apply(classification)
```

```python
regon = pd.read_csv('keywords.csv')
regon[:10]
```

```python
regions = {'Центр': ['москва', 'тула', 'ярославль'],
             'Северо-Запад': ['петербург', 'псков', 'мурманск'],
             'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']}
towns=[]
for values in regions.values():
    for i in values:
        towns.append(i)
towns
```

```python
row='вконтакте моя страница москва'
data=row.split(' ')
def check(t):
    for i in t:
        if i in towns:
            for items in regions.items():
                if i in items[1]:
                    print(items[0])
check(data)
```

```python
def region(row):
    data=row['keyword'].split(' ')
    i=0
    for word in data:
        if word in towns:
            for items in regions.items():
                if word in items[1]:
                    i+=1
                    return items[0]
    if i==0:
        return 'undefined'    
```

```python
regon['region'] = regon.apply(region,axis=1)
regon[:5]
```

```python
regon[(regon['region'] != 'undefined')][:50]
```

```python
years = np.arange(1950,2011)
years  #creat list with years
```

```python
def production_year(title):
  
  for year in years:
    if str(year) in title:
      return year
  return 1900
     
```

```python
movi_rating['year'] = movi_rating['title'].apply(production_year)
movi_rating[100:110]
```

```python
movis = movi_rating.set_index('movieId')
```

```python
movis.groupby('year').agg({'rating':'mean'}).sort_values(by="rating", ascending=False)[:10]
```
