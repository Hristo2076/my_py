```python
log = pd.read_csv('visit_log.csv',sep=';')
log[:5]
```

если источник traffic_source равен Yandex или Google, то в source_type ставится organic;
для источников paid и email из России ставим ad;
для источников paid и email не из России ставим other;
все остальные варианты берём из traffic_source без изменений.


```python
%%time
log.loc[log.traffic_source == ('yandex' and 'google'), 'source_type'] = 'organic'
log.loc[(log.traffic_source == ('paid' and'email')) & (log.region=='Russia'), 'source_type'] = 'ad'
log.loc[(log.traffic_source == ('paid' and'email')) & (log.region!='Russia'), 'source_type'] = 'other'
log.loc[log.source_type.isnull(), 'source_type'] = log['traffic_source']
log[:30]
```

```python
rating = pd.read_csv('ratings.csv',sep=',')
rating[:5]
```

```python
rating.pivot_table(index = 'userId',
                  columns = 'rating', 
                  values = 'timestamp',
                   aggfunc='count',
                   margins=True).head(10)
```

```python
ratings_count = rating.groupby('userId').count().reset_index() #count the number of ratings left by the user
ratings_count=ratings_count[['userId','rating']]
ratings_count[:10] 
```

```python
more_hund = ratings_count[ratings_count.rating > 100] #filte values of more than a hundred
more_hund.rename(columns={'rating': 'rating_count'},inplace=True) 
more_hund.head(10)
```

```python
rating_max=rating.groupby('userId').max().reset_index()
rating_min=rating.groupby('userId').min().reset_index() #group by min/max 

max_time=rating_max[['userId','timestamp']]
min_time=rating_min[['userId','timestamp']]  #leave only the time indicator


max_time.rename(columns={'timestamp': 'timestamp_max'}, inplace=True)
min_time.rename(columns={'timestamp': 'timestamp_min'}, inplace=True) #rename columns

min_time[:10]
```

```python
avg_time= more_hund.merge(max_time, how='inner').merge(min_time, how='inner') #merge min/max time
avg_time[:5]
```

```python
avg_time['avg']=avg_time.apply(lambda row: row['timestamp_max'] - row['timestamp_min'] , axis=1) #calculet avg time 
len(more_hund)==len(avg_time)
```
